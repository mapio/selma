```python
import numpy as np
import matplotlib.pyplot as plt

from selma.geom import *
```

```python
P = np.array([-2, 3])  # red
Q = np.array([2, -4])  # blue

r = -5
c, t, a = compute_arc(P, Q, r)
print(t, a)
circle = plt.Circle(c, r, color='green', fill=False)
plt.gca().add_patch(circle)
plt.scatter(*c, color='black', marker='x')

x = c[0] + abs(r) * np.cos(t)
y = c[1] + abs(r) * np.sin(t)
plt.plot([c[0], x], [c[1], y], color='red')

x = c[0] + abs(r) * np.cos(t + a)
y = c[1] + abs(r) * np.sin(t + a)
plt.plot([c[0], x], [c[1], y], color='blue')


plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.scatter(*P, color='red')
plt.scatter(*Q, color='blue')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xticks([])
plt.yticks([])
plt.gca().set_aspect('equal', adjustable='box')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.show()
```
