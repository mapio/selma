```python
from manim import *
from selma.geom import *
```

```python
%%manim --hide-splash -qm -v WARNING IntArcs

class IntArcs(Scene):
  def construct(self):
      P = Text('P').move_to(3*RIGHT + 2*DOWN)
      Q = Text('Q').move_to(ORIGIN)
      R = Rectangle(width=3, height=1).move_to(P.get_center())
      S = Rectangle(width=2, height=3).move_to(Q.get_center())
      self.add(P, Q, R, S)

      radius = +6
      f, p = get_arcs(R, S, radius)
      p.add_tip(tip_length=0.2).set(color=RED)
      self.add(f, p)
```
